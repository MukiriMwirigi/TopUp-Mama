#### FEATURE SCALING

```python
column_names = ['Is_Blocked', 'Last_Used_Platform', 'Language', 'Country', 'Pick_up_From', 'Category_Name', 'Payment_Type', 'Payment_Method']
features = df[column_names]
sc = StandardScaler().fit(features.values)
features = sc.transform(features.values)
data1 = pd.DataFrame(features, columns = column_names)
data1.head()
```

### MODELING

#### CLUSTERING MODEL

```python
E = []

for cluster in range(1,5):
    kmeans = KMeans( n_clusters = cluster, init='k-means++')
    kmeans.fit(data1)
    E.append(kmeans.inertia_)

# converting the results into a dataframe and plotting them

frame = pd.DataFrame({'Cluster':range(1,5), 'E':E})
```

```python
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['E'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show
```

#### SILHOUETTE COEFFICIENT

A silhouette coefficient, or a silhouette score is a metric used to evaluate the quality of clusters created by the algorithm.


```python
# BUILDING A MODEL WITH 4 CLUSTERS

kmeans = KMeans(n_clusters = 4, init='k-means++')
kmeans.fit(data1)

# The silhouette score of this model

print(silhouette_score(df, kmeans.labels_, metric='euclidean'))
```

```python
#clusters = kmeans.fit_predict(df.iloc[:,1:])
data1["label"] = df.apply(lambda row: kmeans.fit_predict(df.iloc[:,1:]), axis=1)
 
fig = plt.figure(figsize=(21,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data1.Category_Name[data1.label == 0], data1["Payment_Method"][data1.label == 0], df["Country"][data1.label == 0], c='blue', s=60)

ax.scatter(data1.Category_Name[df.label == 1], data1["Payment_Method"][data1.label == 1], data1["Country"][data1.label == 1], c='red', s=60)
ax.scatter(data1.category_Name[df.label == 2], data1["Payment_Method"][data1.label == 2], df["Country"][data1.label == 2], c='green', s=60)
ax.scatter(data1.Category_Name[data1.label == 3], data1["Payment_Method"][data1.label == 3], data1["Country"][data1.label == 3], c='orange', s=60)

ax.view_init(30, 185)
plt.show()
```

#### PCA - TO IMPROVE MODEL USE PCA TO IMPROVE FEATURE SELECTION. 
IT REDUCES THE DIMENSIONALITY AND CREATES NEW COMPONENTS THAT EXPLAIN MAXIMUM VARIANCE OF THE MODEL

```python
pca = PCA(n_components=4)
principalComponents = pca.fit_transform(data1)

features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_ratio_, color='black')
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)

PCA_components = pd.DataFrame(principalComponents)
```

```python
ks = range(1, 10)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(PCA_components.iloc[:,:2])
    inertias.append(model.inertia_)

plt.plot(ks, inertias, '-o', color='black')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

```python
model = KMeans(n_clusters=4)
model.fit(PCA_components.iloc[:,:2])

# silhouette score
print(silhouette_score(PCA_components.iloc[:,:2], model.labels_, metric='euclidean')) 
```

```python
model = KMeans(n_clusters=4)

clusters = model.fit_predict(PCA_components.iloc[:,:2])
data1["label"] = clusters
 
fig = plt.figure(figsize=(21,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data1.Category_Name[data1.label == 0], data1["Payment_Method"][data1.label == 0], data1["Country"][data1.label == 0], c='blue', s=60)
ax.scatter(data1.Category_Name[data1.label == 1], data1["Payment_Method"][data1.label == 1], data1["Country"][data1.label == 1], c='red', s=60)
ax.scatter(data1.Category_Name[data1.label == 2], data1["Payment_Method"][data1.label == 2], data1["Country"][data1.label == 2], c='green', s=60)

ax.scatter(data1.Category_Name[data1.label == 3], data1["Payment_Method"][data1.label == 3], data1["Country"][data1.label == 3], c='orange', s=60)

ax.view_init(30, 185)
plt.show()
```
